### Part 1) 
Implement Ridge and compute parametrization for different values of $\lambda$

```python
import numpy as np
import matplotlib.pyplot as plt



#Generate data
np.random.seed(100)
n = 100
x = np.random.rand(n)
y = 5*x**2 + 0.1*np.random.randn(n)

#design matrix to fit a 2nd order polynomial
X = np.zeros((n,3))
X[:,0] = 1
X[:,1] = x
X[:,2] = x**2


#Rigde regression
lambd = 0.1
L = np.zeros((3,3))

    
def ridge(X, lambd):
    
    for i in range(3):
        L[i,i] = lambd
    
    #find beta vals
    beta_ridge = np.linalg.inv(X.T.dot(X) + L).dot(X.T).dot(y)
    
    #find yvals
    y_tilde = X @ beta_ridge
    
    return y_tilde
    

#get idx of sorted x-values
idx_sorted = np.argsort(x)




#plot result
plt.scatter(x, y, color='r', label='data')
plt.plot(x[idx_sorted], ridge(X, lambd)[idx_sorted], label='model')
plt.legend()
plt.show()



```

### Compare different values of $\lambda$:

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error


lambda_vals = np.linspace(0.1, 1, 10)
mse_vals = np.zeros(len(lambda_vals))
r2_vals = np.zeros(len(lambda_vals))

for i in range(len(lambda_vals)):
    y_ridge = ridge(X, lambda_vals[i])
    mse_vals[i] = mean_squared_error(y, y_ridge)
    r2_vals[i] = r2_score(y, y_ridge)

    
plt.scatter(lambda_vals, mse_vals, label='mean squared error')
plt.xlabel('lambda')
plt.ylabel('MSE')
plt.legend()


```

```python
plt.scatter(lambda_vals, r2_vals, label='r2 score')
plt.xlabel('lambda')
plt.ylabel('R2 score')
plt.legend()
```

MSE increases and R2 decreases as $\lambda$ is increased from 0.1 to 1

## Part 2) 
Repeat the above with scikit-learn

```python
import sklearn.linear_model as skl


_lambda = 0.1
clf_ridge = skl.Ridge(alpha=_lambda).fit(X, y)
y_skl_ridge = clf_ridge.predict(X)

plt.scatter(x, y, label='data')
plt.plot(x[idx_sorted], y_skl_ridge[idx_sorted], label='ridge model')
plt.legend()

print('lambda =', _lambda)
print('R2 score: ', r2_score(y, y_skl_ridge))
print('MSE:', mean_squared_error(y, y_skl_ridge))
```
